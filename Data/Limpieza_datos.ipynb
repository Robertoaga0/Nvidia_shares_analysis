```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nvidia_stock_data.csv")
df
```

```python
print("Valores nulos por columna:\n", df.isnull().sum())
```

```python
#Convertir fechas#

if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])

#Ordenar por fechas#

df.sort_values(by="date", ascending=True, inplace=True)

#eliminar duplicados#

df.drop_duplicates(inplace=True)
```

```python
cols_to_check = ['open', 'high', 'low', 'close', 'adjClose', 'volume']
for col in cols_to_check:
    if col in df.columns:
        print(f"Rango de {col}: Min={df[col].min()}, Max={df[col].max()}")
```

```python
df["date"] = pd.to_datetime(df["date"]).dt.date
```

```python
splits = [
    {"date": "2021-07-20", "factor": 4},
    {"date": "2024-06-10", "factor": 10},
]

# Convert 'date' column to datetime if it's not already
df["date"] = pd.to_datetime(df["date"])

# Adjust prices based on splits
for split in splits:
    df.loc[df["date"] < split["date"], ["close", "high", "low", "open"]] /= split["factor"]
```
